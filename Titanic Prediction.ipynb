```python
def Outlier_Treatment(col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1   
    upper = Q3 + (1.5 * IQR)
    lower = Q1 - (1.5 * IQR)
    np.clip(df[col], lower, upper, inplace = True)
```

```python
Outlier_Treatment('age')
plt.figure(figsize = (10, 7))
sns.boxplot(df['age'])
```

```python
Outlier_Treatment('fare')
plt.figure(figsize = (10, 7))
sns.boxplot(df['fare'])
```
